```python
# Getting this settled

files = ["28BBZ_CD19CarT_Raji_heavy.xls",
         "28BBZ_Raji_CD19CarT_light.xls",
         "28Z_CD19CarT_Raji_light.xls",
         "28Z_Raji_CD19CarT_heavy.xls",
         "28Z_CSF1car_THP1_light.xls",
         "28Z_THP1_CSF1RCar_heavy.xls"]

outfiles = [["28BBZ_CD19CarT_Raji_heavy_2m.pdf", "28BBZ_CD19CarT_Raji_heavy_5m.pdf"],
         ["28BBZ_Raji_CD19CarT_light_2m.pdf", "28BBZ_Raji_CD19CarT_light_5m.pdf"],
         ["28Z_CD19CarT_Raji_light_2m.pdf", "28Z_CD19CarT_Raji_light_5m.pdf"],
         ["28Z_Raji_CD19CarT_heavy_2m.pdf", "28Z_Raji_CD19CarT_heavy_5m.pdf"],
         ["28Z_CSF1car_THP1_light_2m.pdf", "28Z_CSF1car_THP1_light_5m.pdf"],
         ["28Z_THP1_CSF1RCar_heavy_2m.pdf", "28Z_THP1_CSF1RCar_heavy_5m.pdf"]]

exp_names = ["28BB$\zeta$-CAR (CD19) + Raj (Raji pY Sites)",
            "28BB$\zeta$-CAR (CD19) + Raji (CAR pY Sites)",
            "28$\zeta$-CAR (CD19) + Raji (Raji pY Sites)",
            "28$\zeta$-CAR (CD19) + Raji (CAR pY Sites)",
            "28$\zeta$-CAR (CSF1R) + THP-1 (THP-1 pY Sites)",
            "28$\zeta$-CAR (CSF1R) + THP-1 (CAR pY Sites)"]

xaxis_strs = ["$\log_{2}(2m / 0m)$",
              "$\log_{2}(5m / 0m)$"]

yaxis_strs = ["$-\log_{10}(q)$",
              "$-\log_{10}(q)$"]

columns = ["UNIPROT Gene Name", 
           "phosphosite annotated",
           "SILAC ratio 23 for user selected SILAC timepoint1",
           "SILAC ratio 23 for user selected SILAC timepoint2",
           "qvalues for SILAC timepoint1",
            "qvalues for SILAC timepoint2"]

colours = [mph.handle_colours("greens", 3),
           mph.handle_colours("pinks", 3),
           mph.handle_colours("blues", 3), 
           mph.handle_colours("purples", 3),
           mph.handle_colours("browns", 3),
           mph.handle_colours("reds", 3)]
```

```python
def unique_site(q_num, min_q, fcs):
    return None

def replace_nans(q_list, replace = 1):
    newlist = []
    for item in q_list:
        if item == item:
            newlist.append(item)
        else:
            newlist.append(replace)
    return newlist

def filter_df_rows(a_df, column = "U_ID"):
    cols = list(a_df.columns)
    keycol = cols.index(column)
    a_list = list(a_df.to_numpy())
    first_occ = []
    first_key = []
    for row in a_list:
        if row[keycol] not in first_key:
            first_occ.append(row)
            first_key.append(row[keycol])
        else:
            pass
    return pd.DataFrame(first_occ, columns = cols)
    

def read_and_filter(a_file, columns):
    file = pd.read_excel(a_file)
    file = file[columns]
    file["U_ID"] = file[columns[0]] + file[columns[1]]
    fcs = list(file[columns[2:4]].astype(float).to_numpy())
    fcs = [[abs(value) for value in row] for row in fcs]
    qs = list(file[columns[4:]].astype(float).to_numpy())
    qs = [replace_nans(row) for row in qs]
    file["q_num"] = [sum([1 for _ in row if _ < 0.05]) for row in qs]
    file["min_q"] = [min(row) for row in qs]
    file["max_fc"] = [max(row) for row in fcs]
    file = file.sort_values(["U_ID", "q_num", "min_q", "max_fc"],
                            ascending = [True, False, True, False])
    file = filter_df_rows(file)
    return file
```

```python
# Grab the data
file_dfs = [read_and_filter(f, columns) for f in files]

from math import log2

tc = columns[2:]

fcs = np.log2(file_dfs[0][columns[2]].astype(float))
qs = file_dfs[0][columns[4]].astype(float)

mph.volcano(qs, fcs)
```

```python
from math import log10
def find_lims(all_qs,all_fcs):
    qs = gh.unpack_list(all_qs)
    fc = [abs(item) for item in gh.unpack_list(all_fcs) if item == item]
    return round(-log10(min(qs)))+1, round(max(fcs))+1
```

```python
allfcs = [[list(item) for item in list(df[tc[2:]].astype(float).to_numpy())] for df in file_dfs]

allqs = [[list(item) for item in list(df[tc[:2]].astype(float).to_numpy())] for df in file_dfs]


ylim, xlim = find_lims(allqs, allfcs)

xlim = 9
ylim = 6
```

```python
for i in range(len(tc)):
    print(f"{i} : {tc[i]}")

sigs = [0.001,0.01,0.05]   

# Plot the data
j=0
for df in file_dfs:
    c = colours[i]
    for i in range(2):
        if i == 0:
            fc = list(df[tc[0]].astype(float).to_numpy())
            #fc = [log2(item) for item in fc]
            qs = list(df[tc[2]].astype(float).to_numpy())
            merge = list(zip(fc,qs))
            merge = [row for row in merge if all([item == item for item in row])]
            fc, qs = list(zip(*merge))
            mph.volcano(qs, fc, fc_transform = "log2",
                       xlim = xlim, ylim = ylim,
                        colours = colours[j],
                        title = exp_names[j],
                        xlabel = xaxis_strs[i],
                        ylabel = yaxis_strs[i],
                        sig_cutoffs = sigs,
                        fc_cutoff = 1,
                        save = outfiles[j][i])
        elif i == 1:
            fc = list(df[tc[1]].astype(float).to_numpy())
            #fc = [log2(item) for item in fc]
            qs = list(df[tc[3]].astype(float).to_numpy())
            fc, qs = sh.remove_nanpairs(fc, qs)
            mph.volcano(qs, fc, fc_transform = "log2",
                        xlim = xlim, ylim = ylim,
                        colours = colours[j],
                        title = exp_names[j],
                        xlabel = xaxis_strs[i],
                        ylabel = yaxis_strs[i],
                        sig_cutoffs = sigs,
                        fc_cutoff = 1,
                        save = outfiles[j][i])
    j+=1
```
