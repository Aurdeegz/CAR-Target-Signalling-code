```python
def unique_site(q_num, min_q, fcs):
    return None

def replace_nans(q_list, replace = 1):
    newlist = []
    for item in q_list:
        if item == item:
            newlist.append(item)
        else:
            newlist.append(replace)
    return newlist

def filter_df_rows(a_df, column = "U_ID"):
    cols = list(a_df.columns)
    keycol = cols.index(column)
    a_list = list(a_df.to_numpy())
    first_occ = []
    first_key = []
    for row in a_list:
        if row[keycol] not in first_key:
            first_occ.append(row)
            first_key.append(row[keycol])
        else:
            pass
    return pd.DataFrame(first_occ, columns = cols)
    

def read_and_filter(a_file, columns):
    file = pd.read_excel(a_file)
    file = file[columns]
    file["U_ID"] = file[columns[0]] + file[columns[1]]
    fcs = list(file[columns[2:4]].astype(float).to_numpy())
    fcs = [[abs(value) for value in row] for row in fcs]
    qs = list(file[columns[4:-1]].astype(float).to_numpy())
    qs = [replace_nans(row) for row in qs]
    file["q_num"] = [sum([1 for _ in row if _ < 0.05]) for row in qs]
    file["min_q"] = [min(row) for row in qs]
    file["max_fc"] = [max(row) for row in fcs]
    file = file.sort_values(["U_ID", "q_num", "min_q", "max_fc"],
                            ascending = [True, False, True, False])
    file = filter_df_rows(file)
    return file 

from math import log10
def find_lims(all_qs,all_fcs):
    qs = gh.unpack_list(all_qs)
    fc = [abs(item) for item in gh.unpack_list(all_fcs) if item == item]
    return round(-log10(min(qs)))+1, round(max(fc))+1
```

```python
from helpers.general_helpers import unpack_list
def count_list(a_list, neg = False):
    """
    """
    counts = {}
    for item in a_list:
        if item in list(counts.keys()):
            counts[item] += 1
        else:
            counts[item] = 1
    if neg:
        return {key : - value for key, value in counts.items()}
    else:
        return counts

def marry(*dicts, missing = 0):
    """
    """
    if type(missing) == list:
        print(missing)
    all_keys = list(set(unpack_list([list(dictionary.keys()) for dictionary in dicts])))
    new_dict = {key : [] for key in all_keys}
    for key in all_keys:
        for dictionary in dicts:
            if key in list(dictionary.keys()):
                new_dict[key].append(dictionary[key])
            else:
                new_dict[key].append(missing)
    return new_dict

def bin_categorical_data(*lists, missing = 0, neg = False):
    # count the lists
    counts = [count_list(item, neg = neg) for item in lists]
    # put them together
    return marry(*counts, missing = missing)

def process_categories_for_bars(dataframe, qcol_list, fccol_list, category = "KEGG category", 
                                q_threshold = 0.05, fc_threshold = 0, delim = ";"):
    filtered_dfs = [dataframe[dataframe[qcol] < q_threshold] for qcol in qcol_list]
    f_dfs_neg = [filtered_dfs[i][filtered_dfs[i][fccol_list[i]] < - fc_threshold] for i in range(len(fccol_list))]
    f_dfs_pos = [filtered_dfs[i][filtered_dfs[i][fccol_list[i]] > fc_threshold] for i in range(len(fccol_list))] 
    # KEnny will deal with this later
    annotes_neg = [unpack_list([item.split(delim) for item in list(df[category].astype(str)) if item != "nan"]) for df in f_dfs_neg]
    annotes_pos = [unpack_list([item.split(delim) for item in df[category].astype(str) if item != "nan"]) for df in f_dfs_pos]
    negs = bin_categorical_data(*annotes_neg, neg = True)
    poss = bin_categorical_data(*annotes_pos)
    return marry(poss, negs, missing = [0 for _ in range(len(qcol_list))])

def bars_ind(n_items,
             n_groups,
             sep_dist = 0.4,
             transpos = True):
    """
    =================================================================================================
    bars_ind(n_items, n_groups, sep_dist, transpos)
    
    This funciton is used to determine the center positions of all bars used for a bar plot, as
    well as the width of each bar.
    
    =================================================================================================
    Arguments:
    
    n_items   ->  An integer representing the number of categories to be plotted
    n_groups  ->  An integer representing the number of groups per category to be plotted
    sep_dist  ->  A float representing the distance between categories in the plot
    transpos  ->  A boolean determining whether to transpose the output list.
    
    =================================================================================================
    Returns: A list of lists determining the position of bars for each group, and the width
             of all bars in the plot.
    
    =================================================================================================
    """
    # Determine the width of a bar, by subtracting the separation distance from 1
    # and dividing by the number of groups.
    width = (1-sep_dist) / n_groups
    # If there is only one group,
    if n_groups == 1:
        # Then return a list with one sublist, centered
        # on the integers.
        return [[i for i in range(len(n_groups))]]
    # Initialize the output list. This will hold the lists returned.
    output = []
    # Next, loop over the number of categories
    for i in range(n_items):
        # Get the left most position of the interval
        interval = i + sep_dist/2 - 0.5
        # Add equidistant points from the interval to the output
        # Up to the number of groups per category and add that
        # to the outputs
        output.append([interval + j * width for j in range(1,n_groups+1)])
    # If the output must be transposed
    if transpos:
        # Then return the transposed output list and width
        return list(zip(*output)), width
    # Otherwise,
    else:
        # Return the list itself and the width
        return output,
```

```python
rename_cols = {"SILAC ratio 23 for user selected SILAC timepoint1" : "2m fc",
           "SILAC ratio 23 for user selected SILAC timepoint2" : "5m fc",
           "qvalues for SILAC timepoint1" : "2m q",
            "qvalues for SILAC timepoint2" : "5m q",
           "Kegg unique index" : "KEGG"}


file_dfs = [read_and_filter(f, columns_nopub) for f in files if "Published" not in f] + [read_and_filter(f, columns_pub) for f in files if "Published" in f]
file_dfs = [df.rename(columns= rename_cols) for df in file_dfs]
for df in file_dfs:
    df["2m fc"] = np.log2(df["2m fc"])
    df["5m fc"] = np.log2(df["5m fc"])

barput = [process_categories_for_bars(df, ["2m q", "5m q"], ["2m fc", "5m fc"], category = "KEGG", delim="\n")
          for df in file_dfs]

def sort_bar_values(updown):
    upmax = max(updown[0])
    downmax = -min(updown[1])
    return max([upmax, downmax])

```

```python
def set_ylim(ax, lim = 0):
    if not bool(lim):
        ylim = max([abs(y) for y in ax.get_ylim()])
        ax.set_ylim(-ylim, ylim)
    else:
        ax.set_ylim(-lim, lim)
    return None


outkegg = ["outputs/28bbz_raji_heavy_kegg.pdf",
           "outputs/28bbz_raji_light_kegg.pdf",
           "outputs/28z_raji_light_kegg.pdf",
           "outputs/28z_raji_heavy_kegg.pdf",
           "outputs/28z_thp1_light_kegg.pdf",
           "outputs/28z_thp1_heavy_kegg.pdf",
           "outputs/28bbz_skbr3_heavy_kegg.pdf",
           "outputs/28bbz_skbr3_light_kegg.pdf",
           "outputs/published_raji_kegg.pdf",
           "outputs/published_tcell_kegg.pdf"]

i = 0
for bars in barput:
    keep = [[key, value] for key, value in bars.items()]
    keep = sorted(keep, key = lambda x: -sort_bar_values(x[1]))[:7]
    keep_labs = [cat[0] for cat in keep]
    keep_vals_pos = list(zip(*[cat[1][0] for cat in keep]))
    keep_vals_neg = list(zip(*[cat[1][1] for cat in keep]))
    
    fig, ax = plt.subplots()
    xs = bars_ind(7,2)
    
    labs = ["2m vs 0m", "5m vs 0m"]
    
    for j in range(2):
        ax.bar(xs[0][j], keep_vals_pos[j], color = colours[i][j], edgecolor = "black", width = xs[1], label = labs[j])
        ax.bar(xs[0][j], keep_vals_neg[j], color = colours[i][j], edgecolor = "black", width = xs[1])
    ax.plot([-1,7], [0,0], color = "black")
    ax.set_xticks(xs[0][1])
    ax.set_xticklabels(keep_labs, ha = "right", rotation = 28.7,fontfamily = "sans-serif", font = "Arial", fontweight = "bold")
    set_ylim(ax)

    ax.set_yticks(ax.get_yticks()[1:-1])
    ax.set_yticklabels([str(abs(int(item))) for item in ax.get_yticks()],fontfamily = "sans-serif", font = "Arial", fontweight = "bold")

    fig.text( 0.025, 0.88, fr"Fold change$>1$", rotation = 90, va = "top",fontfamily = "sans-serif", font = "Arial", fontweight = "bold")
    fig.text( 0.025, 0.48, fr"Fold change$<1$", rotation = 90, va = "top",fontfamily = "sans-serif", font = "Arial", fontweight = "bold")
    fig.text(-0.05, 0.5, fr"$\#$ of Unique Sites $(q<0.05)$", rotation = 90, va = "center",fontfamily = "sans-serif", font = "Arial", fontweight = "bold")

    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)
    
    ax.set_title(exp_names[i], fontfamily = "sans-serif", font = "Arial", fontweight = "bold", fontsize = 16)
    ax.legend(loc = "lower right")
    plt.savefig(outkegg[i])
    
    i+=1
```
